## Problem Description:
   - there are an significant need for many companies an banks that specialize  in providing home loans. 
   - in this dataswet the company operate in diverse areas, including urban, semi-urban, and rural regions.
   - When a customer applies for a home loan, the company needs to assess the customer's eligibility for the loan.
   - the eligiblity for the loan based on various details provided in the online application form. 
   - The customer(the applicant for the loan) details include:
         . Gender      . Marital              . Status
         . Education   . Number of Dependent  . Income
         . Loan Amount . Credit History       . Others
   - the main purpose is To streamline and automate the loan eligibility process in real-time,
   - The goal is to identify specific customer segments that are eligible for a loan amount.
   - By doing so, the company aims to target these eligible customers more efficiently.
___       
## About the dataset 
   - this dataset contains information about loan eligibility
   - we can describe the dataset as folows :
   - it contains 614 rows
   - it contains 13 columns
   - the columns are as follows :
        1.  **Loan ID** : Unique identifier for each loan.(its not important for our analysis ).
        2.  **Gender**: Categorical variable indicating the gender of the loan applicant. 
        3.  **Married**: Categorical variable indicating the marital status of the loan applicant.
        4.  **Dependents**: Numeric variable representing the number of dependents the loan applicant has.
        5.  **Education**: Categorical variable indicating the educational level of the loan applicant.
        6.  **Self Employed**: indicated whether the loan applicant is self-employed. (y/n)
        7.  **ApplicantIncome**: Numeric variable representing the income of the loan applicant.
        8.  **Co applicant Income**: Numeric variable representing the income of the co-applicant (if any).
        9.  **Loan Amount**: Numeric variable representing the loan amount.
        10. **Loan Amount Term**: Numeric variable representing the term of the loan in months.
        11. **Credit History**: indicate whether the credit history of the applicant meets the guidelines.
        12. **Property Area**:indicate the area where the property is located. can be Urban, Semi-Urban ,Rural.
        13. **Loan_Status**: variable indicate whether the loan was approved. Y (approved) or N (not approved).

## Decision Tree Classifier

````python
# Initialize the Decision Tree Classifier with possible parameters for tuning
parameters = {
    'max_depth': [7,14,8,10,16,20],
    'min_samples_split': [10,15,12],
    'min_samples_leaf': [5,8,10,12,1],
    'criterion': ['gini', 'entropy']
}

o_DT_grid_search = GridSearchCV(DecisionTreeClassifier(criterion='entropy'), parameters,scoring='f1', cv=5)
o_DT_grid_search.fit(X_train, y_train)
````
